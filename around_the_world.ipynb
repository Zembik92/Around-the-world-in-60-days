## 3) Build a Dictionary of the words used in the book 
- Build a Python dictionary (`dict`) in which the keys are the unique words in the book and the values are dictionary objects with fields `length` for word length and `freq` for frequency, e.g.:

`{'the': {'length': 3, 'freq': 4303}}`


```python
# The code creates a nested dictionary with x as the key and the length and frequency of x as the values, by
# getting the length of x using the len() function and assigning the variable name "length" to it, 
# then getting the frequency of x in book_list using the .count() function and assigning the varible name "freq" to it. 
# The for statement iterates over book_list in order, executing the code each time (meaning for every word in book_list, take the word and apply the above to it).
dict3 = {x:{"length": len(x),"freq":book_list.count(x)} for x in book_list}
print(dict3)                                                                          
```

```python
# Loop over dict3, saving each frequency that is higher than the previous to the variable most_freq untill its left with the highest 
# If the highest value I get is more than 1, I then want it to return the value 
# Else it should return 1
most_freq = 1
for x in dict3:
    if dict3[x]["freq"]> most_freq: 
        most_freq = dict3[x]["freq"]
        first = x
    else:
        most_freq = most_freq
print(first, most_freq) 
```

## 5) Write a generator function to produce each word and its length each time it yields a value 

- The function should accept a single argument: the list of words in the book
- It should yield a tuple (word, word_len)
- Call the function to create a generator but do not iterate over it at this stage

```python
def my_gen(book_list): 
    """I defined a generator my_gen with book_list as an argument,
    
    then iterate through book_list to get the element and the length of the element.
    
    Yields:
    
    I use the yield function to excecute values at a time as required.""" 
    
    for x in book_list:
            yield (x, len(x)) 
print(my_gen(book_list))   
```

## 6) Use a comprehension over the generator you just made to list all the words with 9 letters 
- Print all those words
- Extra points if each word is selected only once

```python
# I iterate over the generator function "my_gen(book_list" looking for two values x (the word) and y(the length of the word)
# The code only returns x if the .isalpha() function criteria that checks whether the string contains alphabetic characters only is met.
# It ensures that words with numbers and punctuations are excluded and then only returns x if y is equal to 9.
# I remove duplicate values by appending x to the list "nine" only if it is not already found in the list.  
nine = []
[nine.append(x) for (x, y) in my_gen(book_list) if (x.isalpha()) if y == 9 if x not in nine]
print(nine)
```

One of my inhebitors was to find an online source which contained all the cities mentioned in the book, different sources can potentially be used to get the required data but this would cause inconsistancy.

```python
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get("https://www.which.co.uk/reviews/destinations/article/best-city-breaks-in-the-uk") 

# Checking if the page was downloaded  successfully. A status_code of 200 means the page downloaded successfully
page.status_code 

# Parse the webpage to convert data into a usable formate, then use .find id = "" to return the table table content we intend to extract   
# I used prettify to display the information in a readable formate
hope = BeautifulSoup(page.content, 'html.parser')
reviews = hope.find(id="leisure_tables")
more = reviews.find("table") 
print(more.prettify())
```

```python
# From our table, I then use find_all to return all the table rows and give it a variable name
table_rows = more.find_all("tr") 
print(table_rows)
```

```python
# Using the table rows found above, I use find_all to return the table data from each table row 
# I then use a comprehension to get only text from the table data by using .text (thats why my table does not have any data in column C-I, there was image data)
# Finally I convert the data to a dataframe and gave it column names as described below   
l = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.text for tr in td]
    l.append(row)
b = pd.DataFrame(l, columns=["City", "Avg hotel rate", "C","D","E","F","City score"])    
b
```

```python
# I used .drop to eliminate the specified rows and I use [[]] because I am eliminating two indices
# .index specifies the rows at index[0] and [1] that i want deleted
m = b.drop(b.index[[0,1]]) 
                        
# Axis=1 denotes that I am referring to a column, not a row and inplace = True is used to affect the drop
m.drop(["C","D","E","F",], axis = 1, inplace = True)
m
```
